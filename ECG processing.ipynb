```python
import scipy.io
from matplotlib import pyplot as plt

mat = scipy.io.loadmat('file.mat')
y=mat['y15']
plt.plot(y)

index_min=y.argmax()
index_value = max(y)
print('ind: ' + str(index_min))
print('value: ' + str(index_value))

#plt.plot(index_min,index_value,color="green", linewidth=1.0, linestyle="*")
plt.annotate(r'$max(y)$',
            xy=(index_min, index_value), xycoords='data',
            xytext=(+10, +30), textcoords='offset points', fontsize=16,
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
```
